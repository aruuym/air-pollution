# Analysis for Air Pollution




###  Abstract 


Air pollution is the main cause of a global environmental threat. The International Labor Organization defines air pollution as the presence in the air of substances that are harmful to health or otherwise dangerous, regardless of their physical form. Burning fossil fuels, farming and mining are just some of the causes of air pollution. Most often and most of all, they pollute the atmosphere: carbon dioxide, sulfur dioxide, nitrogen oxides and dust.


### Introduction

Polluted air is absorbed by humans mainly through breathing. It promotes the development of respiratory diseases, allergies, and reproductive disorders. In human life, air pollution causes corrosion of metals and building materials. It also negatively affects the plant world, disrupting the processes of photosynthesis, transpiration and respiration. Polluted air also degrades the condition of water and soil. Globally, air pollution has an impact on climate change. Air pollution also increases the acidity of drinking water. This causes an increase in the content of lead, copper, zinc, aluminum and even cadmium in the water entering our apartments. Highly acidic water destroys plumbing systems, flushing out various toxic substances from them.

# Q1: Which countries in red zone and struggling from air pollution.

```python
var2 = var2[['Deaths - Air pollution - Sex: Both - Age: All Ages (Number)','Entity']]
var2
```

```python
sdf = var2.groupby(['Entity']).sum().reset_index().sort_values(by='Deaths - Air pollution - Sex: Both - Age: All Ages (Number)', ascending=False)
sdf
```

```python
fig, ax = plt.subplots(figsize=(20,5))
sns.barplot(data=sdf.head(20), x = 'Entity', y = 'Deaths - Air pollution - Sex: Both - Age: All Ages (Number)', ax=ax)
plt.xticks(rotation=45, ha='right');
```

```python
sdf2 = sdf[:15]
sdf2
```

```python
new_row = pd.DataFrame(data = {
    'Entity' : ['others'],
    'Deaths - Air pollution - Sex: Both - Age: All Ages (Number)' : [sdf['Deaths - Air pollution - Sex: Both - Age: All Ages (Number)'][15:].sum()]
})
new_row
```

```python
sdf2 = pd.concat([sdf2, new_row])
sdf2
```

```python
plt.subplots(figsize=(20,10))
sdf2.groupby(['Entity'])['Deaths - Air pollution - Sex: Both - Age: All Ages (Number)'].sum().sort_values().plot.pie(autopct="%.1f%%")
plt.ylabel(ylabel='')
plt.suptitle('Deaths - Air pollution - Sex: Both - Age: All Ages (Number)', fontsize=16)
```

# Q4: Number and type of monitoring stations

```python
sdf = var[['Number and type of monitoring stations','Region']]
sdf = sdf.rename(columns={'Number and type of monitoring stations':'number'})
sdf
```

```python
sdf['number'] = sdf['number'].str.split(n=1).str[1]
sdf
```

```python
sdf = (sdf.set_index(['Region'])
   .apply(lambda x: x.str.split(',').explode())
   .reset_index())  
sdf = sdf[sdf['number']!='']
sdf['hhh'] = 'gg'
sdf
```

```python
sdf = sdf.groupby(['number', 'Region']).count().reset_index().sort_values(by='hhh', ascending=False)
sdf
```

```python
sdf = sdf.groupby(['number'])['hhh'].sum().reset_index().sort_values(by='hhh', ascending=False)
sdf
```

```python
fig, ax = plt.subplots(figsize=(20,5))
sns.barplot(data=sdf.head(15), x = 'number', y = 'hhh', ax=ax)
plt.xticks(rotation=45, ha='right');
```
